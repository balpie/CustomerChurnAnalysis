```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
df = pd.read_csv("telco-customer-churn.csv")
```

```python
print(list(df.columns))
```

```python
#Name: gender, Length: 7043, dtype: str
MaleCounter = 0
FemaleCounter = 0
for elem in df['gender']:
    if elem == 'Male':
        MaleCounter = MaleCounter + 1
    if elem == 'Female':
        FemaleCounter = FemaleCounter + 1
print("FemaleCounter: ",FemaleCounter)  #3488
print("MaleCounter: ",MaleCounter)      #3555
print(MaleCounter+FemaleCounter)
```

```python
#Name: SeniorCitizen, Length: 7043, dtype: int64
#print(df['SeniorCitizen'])
SeniorCounter = 0
NotSeniorCounter = 0
for elem in df['SeniorCitizen']:
    if elem == 1:
        SeniorCounter = SeniorCounter + 1
    if elem == 0:
        NotSeniorCounter = NotSeniorCounter + 1
print("SeniorCounter: ",SeniorCounter)          #1142
print("NotSeniorCounter: ",NotSeniorCounter)    #5901
print(SeniorCounter+NotSeniorCounter)
```

```python
#print(df['Partner'])
#Name: Partner, Length: 7043, dtype: str

Partner = 0
NotPartner = 0
for elem in df['Partner']:
    if elem == 'Yes':
        Partner = Partner + 1
    if elem == 'No':
        NotPartner = NotPartner + 1
print("Partner: ",Partner)         #3402 
print("NotPartner: ",NotPartner)   #3641  
print(Partner+NotPartner)
```

```python
print(df['Dependents'])
#Name: Dependents, Length: 7043, dtype: str

Dependents = 0
NotDependents = 0
for elem in df['Dependents']:
    if elem == 'Yes':
        Dependents = Dependents + 1
    if elem == 'No':
        NotDependents = NotDependents + 1
print("Dependents: ",Dependents)         #2110 
print("NotDependents: ",NotDependents)   #4933 
print(Dependents+NotDependents)
```

```python
#print(df['tenure'])
#Number of months the customer has stayed with the company
tenuremax = df['tenure'].max()
tenuremin = df['tenure'].min()
print("tenuremax: ", tenuremax ,"tenuremin: ", tenuremin)

#distribuzione della tenure
sns.histplot(df['tenure'],  bins=100)
plt.title('Distribuzione di tenure')
plt.show()

#plotting tenure e churn
sns.scatterplot(
    data= df,
    x='tenure',
    y='MonthlyCharges',
    hue='Churn',
    palette='viridis'
)

plt.title('Grafico')
plt.show()




```
